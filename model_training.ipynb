Here we will use YOLO Algorithm for building classifier. And we will use yolovm.pt pretrained weights from ultralytics.

### Required modules

```python
# required modules
from ultralytics import YOLO
```

### Loading pretrained weights

```python
# loading pretrained weights
model = YOLO('yolov8m.pt')
```

### Model Training

Here we uses
- data = data.yaml -----> where custom data are stored
- epochs = 100 -----> Number of epochs in training
- imgs = 640 -----> image size

```python
results = model.train(data='data.yaml', epochs=100, imgsz=640)
```
